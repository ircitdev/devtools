## 3. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

````python
# –î–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤ notebook
%matplotlib inline

import matplotlib.pyplot as plt
````

````python
# –ü—Ä–æ—Å—Ç–æ–π –≥—Ä–∞—Ñ–∏–∫
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ sin(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````

````python
# –°—Ç–æ–ª–±—á–∞—Ç–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞
plt.figure(figsize=(8, 5))
df.plot(x='–ò–º—è', y='–í–æ–∑—Ä–∞—Å—Ç', kind='bar', color='skyblue')
plt.title('–í–æ–∑—Ä–∞—Å—Ç —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤')
plt.xlabel('–ò–º—è')
plt.ylabel('–í–æ–∑—Ä–∞—Å—Ç')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````
